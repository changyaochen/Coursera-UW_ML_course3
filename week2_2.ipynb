```python
import pandas as pd
import numpy as np
import json, string
```

```python
products = pd.read_csv('./amazon_baby_subset.csv')
products['review'].fillna('', inplace=True)
with open('important_words.json', 'r') as f:
    important_words = json.load(f)
```

```python
def remove_punctuation(text):
    table = str.maketrans({key: None for key in string.punctuation})
    return str(text).translate(table)

products['review_clean'] = products['review'].apply(remove_punctuation)
# add the word count as features
for word in important_words:
    products[word] = products['review_clean'].apply(lambda s : s.split().count(word))
```

```python
table = pd.DataFrame({'word': ['(intercept)'] + important_words})
```

```python
with open('module-4-assignment-train-idx.json', 'r') as f:
    train_idx = json.load(f)
with open('module-4-assignment-validation-idx.json', 'r') as f:
    valid_idx = json.load(f)

train_data = products.iloc[train_idx, :].reset_index(drop=True)
validation_data = products.iloc[valid_idx, :].reset_index(drop=True)
```

```python
def get_X_and_y(df, feature_list, label_col):
    X = np.insert(df[feature_list].values, 0, 1, axis=1)    
    y = df[label_col].values
    return X, y
```

```python
feature_matrix_train, sentiment_train = get_X_and_y(train_data, important_words, 'sentiment')
feature_matrix_valid, sentiment_valid = get_X_and_y(validation_data, important_words, 'sentiment') 
```

```python
def predict_probability(feature_matrix, coefficients):
    assert(len(coefficients) == feature_matrix.shape[1]) 
    return 1./(1+np.exp(-1.*np.dot(feature_matrix, coefficients)))

def feature_derivative(errors, feature):
    return np.dot(errors, feature)

def compute_log_likelihood(feature_matrix, sentiment, coefficients, lambda_):
    indicator = (sentiment==+1)
    scores = np.dot(feature_matrix, coefficients)
    lp = np.sum((indicator-1)*scores - np.log(1. + np.exp(-scores))) \
        - lambda_*np.square(coefficients[1:]).sum()
    return lp

def logistic_regression(feature_matrix, sentiment, initial_coefficients, 
                        step_size, max_iter, lambda_, verbose=False):
    coefficients = initial_coefficients
    for itr in range(max_iter):
        proba = predict_probability(feature_matrix, coefficients)
        indicator = (sentiment==1)
        errors = indicator - proba
        derivative = feature_derivative(errors, feature_matrix)
        coefficients += step_size*(derivative - 2*lambda_*np.insert(coefficients[1:], 0, 0))
        if (itr <= 15 or (itr <= 100 and itr % 10 == 0) or (itr <= 1000 and itr % 100 == 0) 
            or (itr <= 10000 and itr % 1000 == 0) or itr % 10000 == 0) \
            and verbose:
            lp = compute_log_likelihood(feature_matrix, sentiment, coefficients, lambda_)
            print('iteration %*d: log likelihood of observed labels = %.8f' % 
                (int(np.ceil(np.log10(max_iter))), itr, lp))
    return coefficients  
```

```python
lambda_s = [0, 4, 10, 1e2, 1e3, 1e5]
coef_s = []
for lambda_ in lambda_s:
    coef_ = logistic_regression(feature_matrix_train, sentiment_train,
                            np.zeros(feature_matrix_train.shape[1]), 5e-6, 501, lambda_)
    table[str(lambda_)] = coef_
    coef_s.append(coef_)
```

```python
positive_words = np.argsort(coef_s[0])[-5:]
table.iloc[positive_words, :]
```

```python
negative_words = np.argsort(coef_s[0])[:5]
table.iloc[negative_words, :]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 6

def make_coefficient_plot(table, positive_words, negative_words, l2_penalty_list):
    cmap_positive = plt.get_cmap('Reds')
    cmap_negative = plt.get_cmap('Blues')
    
    xx = l2_penalty_list
    plt.plot(xx, [0.]*len(xx), '--', lw=1, color='k')
    
    table_positive_words = table.iloc[positive_words, :]
    table_negative_words = table.iloc[negative_words, :]
    
    for i in range(len(positive_words)):
        color = cmap_positive(0.8*((i+1)/(len(positive_words)*1.2)+0.15))
        plt.plot(xx, table_positive_words.iloc[i:i+1, 1:].as_matrix().flatten(),
                 '-', label=table_positive_words.iloc[i, 0], linewidth=4.0, color=color)
        
    for i in range(len(negative_words)):
        color = cmap_negative(0.8*((i+1)/(len(negative_words)*1.2)+0.15))
        plt.plot(xx, table_negative_words.iloc[i:i+1, 1:].as_matrix().flatten(),
                 '-', label=table_negative_words.iloc[i, 0], linewidth=4.0, color=color)
        
    plt.legend(loc='best', ncol=3, prop={'size':16}, columnspacing=0.5)
    plt.axis([1, 1e5, -1, 2])
    plt.title('Coefficient path')
    plt.xlabel('L2 penalty ($\lambda$)')
    plt.ylabel('Coefficient value')
    plt.xscale('log')
    plt.rcParams.update({'font.size': 18})
    plt.tight_layout()


make_coefficient_plot(table, positive_words, negative_words, 
                      l2_penalty_list=[0, 4, 10, 1e2, 1e3, 1e5])
```

```python
for i, coef_ in enumerate(coef_s):
    print("\nL2 coefficient is: ", lambda_s[i])
    proba = predict_probability(feature_matrix_train, coef_)
    pred = (proba > 0.5).astype(int)
    target = [1 if e == 1 else 0 for e in sentiment_train]
    accu = (pred == target).sum() / pred.shape[0]
    print("Accuracy on training set: {0:6.5f}".format(accu))
    
    proba = predict_probability(feature_matrix_valid, coef_)
    pred = (proba > 0.5).astype(int)
    target = [1 if e == 1 else 0 for e in sentiment_valid]
    accu = (pred == target).sum() / pred.shape[0]
    print("Accuracy on validation set: {0:6.5f}".format(accu))
```
